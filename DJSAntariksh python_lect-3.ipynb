```python
#Matplot-lib

#1 : line graph

import matplotlib.pyplot as plt 
x = [1,2,3] 
y = [2,4,1] 
plt.plot(x, y) 
plt.xlabel('x - axis') 
plt.ylabel('y - axis') 
plt.show() 
```

```python
#2 : double line graph

from matplotlib import style
 
style.use('ggplot')
x = [5,8,10]
y = [12,16,6]
x2 = [6,9,11]
y2 = [6,15,7]
plt.plot(x,y,'g',label='line one', linewidth=5)
plt.plot(x2,y2,'c',label='line two',linewidth=5)
plt.title('Graph')
plt.ylabel('Y axis')
plt.xlabel('X axis')
plt.legend()
plt.grid(True,color='k')
plt.show()
```

```python
#3 : bar-chart

plt.bar([0.25,1.25,2.25,3.25,4.25],[50,40,70,80,20],
label="BMW",color='b',width=.5)
plt.bar([.75,1.75,2.75,3.75,4.75],[80,20,20,50,60],
label="Audi", color='r',width=.5)
plt.legend()
plt.xlabel('Days')
plt.ylabel('Distance (kms)')
plt.title('Cars')
plt.show()
```

```python
#4 : scatter plot

x = [1,1.5,2,2.5,3,3.5,3.6]
y = [7.5,8,8.5,9,9.5,10,10.5]
 
x1=[8,8.5,9,9.5,10,10.5,11]
y1=[3,3.5,3.7,4,4.5,5,5.2]
 
plt.scatter(x,y, label='high income low saving',color='r')
plt.scatter(x1,y1,label='low income high savings',color='b')
plt.xlabel('saving*100')
plt.ylabel('income*1000')
plt.title('Scatter Plot')
plt.legend()
plt.show()
```

```python
# Panda Data-Frames

#1:
import pandas as pd

data = [['John', 25, 'New York'],
       ['Alice', 30, 'London'],
       ['Bob', 35, 'Paris']]

df = pd.DataFrame(data, columns=['Name', 'Age', 'City'])

print(df)

mean_age = df['Age'].mean()
print(f"Mean Age: {mean_age}")

aggregate_stats = df['Age'].describe()
print("\nAggregate Statistics:")
print(aggregate_stats)

age_key_value_pairs = df.set_index('Name')['Age'].to_dict()
print("\nKey-Value Pairs (Name: Age):")
print(age_key_value_pairs)

```

```python
#2 :

info = {'ID' :[101, 102, 103],'Department' :['B.Sc','B.Tech','M.Tech',]}  
df = pd.DataFrame(info)   
df['Salary'] = [50000, 60000, 75000]  

print(df)

mean_salary = df['Salary'].mean()
print(f"Mean Salary: {mean_salary}")

aggregate_stats_salary = df['Salary'].describe()
print("\nAggregate Statistics for Salary:")
print(aggregate_stats_salary)

id_salary_pairs = df.set_index('ID')['Salary'].to_dict()
print("\nKey-Value Pairs (ID: Salary):")
print(id_salary_pairs)
```
