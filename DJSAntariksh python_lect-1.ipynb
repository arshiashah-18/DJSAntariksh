```python
# strings

#1: Frequency of Characters

def character_frequency(s):
    freq = {}
    for char in s:
        if char in freq:
            freq[char] += 1
        else:
            freq[char] = 1
    return freq

string = "hello world"
print(f"Character frequency in '{string}': {character_frequency(string)}")


#2: check if a substring is present in a string
string = input("Enter a string : ")
string2 = input("enter the word to be found : ")

if string2 in string:
    print("yes")
else :
    print("no")

#3 : String Compression

def compress_string(s):
    compressed = ""
    count = 1
    for i in range(1, len(s)):
        if s[i] == s[i - 1]:
            count += 1
        else:
            compressed += s[i - 1] + str(count)
            count = 1
    compressed += s[-1] + str(count)
    return compressed if len(compressed) < len(s) else s

string = "aabcccccaaa"
print(f"Compressed string of '{string}': {compress_string(string)}")

#4 : Count Substrings

def count_substrings(s, substring):
    count = 0
    start = 0
    while True:
        start = s.find(substring, start) + 1
        if start > 0:
            count += 1
        else:
            break
    return count

string = "hello hello hello world"
substring = "hello"
print(f"Number of occurrences of '{substring}' in '{string}': {count_substrings(string, substring)}")


#5 : number of words in a sentence

str = input("Enter a sentence")
result = len(str.split())
print(result)
```

```python
# Numbers

#1 : Perfect Number Checker

def is_perfect_number(num):
    if num <= 1:
        return False
    divisors_sum = sum(i for i in range(1, num) if num % i == 0)
    return divisors_sum == num

number = 28
print(f"{number} is a perfect number: {is_perfect_number(number)}")

#2 : Sum of Multiples

def sum_of_multiples(n1, n2, limit):
    sum_multiples = 0
    for num in range(1, limit):
        if num % n1 == 0 or num % n2 == 0:
            sum_multiples += num
    return sum_multiples

number1 = 3
number2 = 5
range_limit = 1000
result = sum_of_multiples(number1, number2, range_limit)
print(f"The sum of multiples of {number1} and {number2} below {range_limit} is: {result}")


#3 : Binary to Decimal Conversion

def binary_to_decimal(binary):
    decimal = 0
    power = len(binary) - 1
    for digit in binary:
        decimal += int(digit) * (2 ** power)
        power -= 1
    return decimal

binary_number = "101010"
print(f"Decimal equivalent of {binary_number}: {binary_to_decimal(binary_number)}")

#4 : Fibonacci Sequence Generator

def fibonacci_sequence(n):
    fib_sequence = [0, 1]
    while len(fib_sequence) < n:
        fib_sequence.append(fib_sequence[-1] + fib_sequence[-2])
    return fib_sequence

terms = 10
print(f"Fibonacci sequence with {terms} terms: {fibonacci_sequence(terms)}")

#5 : Decimal to roman number conversion

def printRoman(number):
    num = [1, 4, 5, 9, 10, 40, 50, 90,
        100, 400, 500, 900, 1000]
    sym = ["I", "IV", "V", "IX", "X", "XL",
        "L", "XC", "C", "CD", "D", "CM", "M"]
    i = 12
     
    while number:
        div = number // num[i]
        number %= num[i]
 
        while div:
            print(sym[i], end = "")
            div -= 1
        i -= 1
 
if __name__ == "__main__":
    number = 3549
    print("Roman value is:", end = " ")
    printRoman(number)
```

```python
# dictionary

#1 : Merging 2 Dictionaries

def Merge(dict1, dict2):
    return(dict2.update(dict1))

dict1 = {'a': 10, 'b': 8}
dict2 = {'d': 6, 'c': 4}

print(Merge(dict1, dict2))

print(dict2)

#2 : Add Items to a Dictionary

country_capitals = {
  "Germany": "Berlin", 
  "Canada": "Ottawa", 
}

country_capitals["Italy"] = "Rome"

print(country_capitals)

#3 : Sorting dictionary by key or value

myDict = {'ravi': 10, 'rajnish': 9,
        'sanjeev': 15, 'yash': 2, 'suraj': 32}

myKeys = list(myDict.keys())
myKeys.sort()
sorted_dict = {i: myDict[i] for i in myKeys}

print(sorted_dict)

#4 : Sum of all elements in dictionary

def returnSum(myDict):
 
    list = []
    for i in myDict:
        list.append(myDict[i])
    final = sum(list)
 
    return final
 
dict = {'a': 100, 'b': 200, 'c': 300}
print("Sum :", returnSum(dict))

#5 : Dictionary Filtering

def filter_dictionary(dictionary, condition):
    filtered_dict = {}
    if condition == 'key_starts_with_a':
        filtered_dict = {k: v for k, v in dictionary.items() if k.startswith('a')}
    elif condition == 'value_greater_than_10':
        filtered_dict = {k: v for k, v in dictionary.items() if v > 10}
    return filtered_dict

sample_dict = {'apple': 5, 'banana': 15, 'avocado': 10, 'orange': 8}
filtered_dict = filter_dictionary(sample_dict, 'value_greater_than_10')
print("Filtered Dictionary:")
print(filtered_dict)
```

```python
# for loop :

#1 : Print Pascal's triangle

from math import factorial


n = 5
for i in range(n):
    for j in range(n-i+1):

        print(end=" ")

    for j in range(i+1):

        print(factorial(i)//(factorial(j)*factorial(i-j)), end=" ")

  
    print()

#2 : Selection Sort

def selection_sort(arr):
    n = len(arr)
    for i in range(n - 1):
        min_index = i
        for j in range(i + 1, n):
            if arr[j] < arr[min_index]:
                min_index = j
        if min_index != i:
            arr[i], arr[min_index] = arr[min_index], arr[i]

array = [64, 25, 12, 22, 11]
print(f"Original array: {array}")
selection_sort(array)
print(f"Sorted array using Selection Sort: {array}")


#3 : Count occurrences of each word in a sentence using a for loop:

def count_words(sentence):
    words = sentence.split()
    word_count = {}
    for word in words:
        if word in word_count:
            word_count[word] += 1
        else:
            word_count[word] = 1
    return word_count

text = "This is a sentence with several words. This sentence is simple but informative."
word_counts = count_words(text)
print("Word counts:", word_counts)

#4 : Factorial Calculation

def factorial(n):
    if n < 0:
        return None  
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

number = 5
print(f"Factorial of {number}: {factorial(number)}")


#5 : Generate all permutations of a list using recursion and loops:

def permutations(lst):
    if len(lst) <= 1:
        return [lst]
    perms = []
    for i in range(len(lst)):
        rest = lst[:i] + lst[i+1:]
        for p in permutations(rest):
            perms.append([lst[i]] + p)
    return perms

elements = [1, 2, 3]  
all_perms = permutations(elements)
print(f"All permutations of {elements}:")
for perm in all_perms:
    print(perm)


```

```python
#Sets

#1 : Remove Duplicates from a List using a Set:

a_list = [1, 2, 3, 4, 2, 3, 5]

unique_elements = set(a_list)
print("List with duplicates removed:", list(unique_elements))

#2 : Finding Common Elements in Multiple Sets:

set1 = {1, 2, 3, 4, 5}
set2 = {3, 4, 5, 6, 7}
set3 = {5, 6, 7, 8, 9}

common_elements = set1 & set2 & set3
print("Common elements in set1, set2, and set3:", common_elements)

#3 : Counting Element Frequencies using a Set and Dictionary:

a_list = [1, 2, 3, 2, 3, 4, 5, 1, 3, 4]

unique_elements = set(a_list)
frequency_dict = {element: a_list.count(element) for element in unique_elements}

print("Element frequencies:")
for element, count in frequency_dict.items():
    print(f"{element}: {count}")

#4 : Checking for Anagrams using Sets:

def are_anagrams(str1, str2):
    return set(str1) == set(str2)

word1 = "listen"
word2 = "silent"
if are_anagrams(word1, word2):
    print(f"{word1} and {word2} are anagrams.")
else:
    print(f"{word1} and {word2} are not anagrams.")

#5 : Finding Common Characters in List of Strings using Sets:

def common_characters(str_list):
    if not str_list:
        return set()
    
    common_chars = set(str_list[0])
    for s in str_list[1:]:
        common_chars &= set(s)
    
    return common_chars

strings = ["flower", "flow", "flight"]
common_chars = common_characters(strings)
print("Common characters in", strings, ":", common_chars)
```

```python
# while loops

#1 : Binary Search Algorithm

def binary_search(arr, target):
    low, high = 0, len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1  

sorted_list = [2, 5, 8, 12, 16, 23, 38, 56, 72, 91]
target = 23
print(f"Index of {target} in the list: {binary_search(sorted_list, target)}")

#3 : Calculate the sum of numbers until user enters 0

number = int(input('Enter a number: '))

total = 0

while number != 0:
    total += number
    number = int(input('Enter a number: '))

print('The sum is', total)

#4 : Fibonacci sequence using a while loop:

def fibonacci(n):
    fib_sequence = []
    a, b = 0, 1
    while len(fib_sequence) < n:
        fib_sequence.append(a)
        a, b = b, a + b
    return fib_sequence

count = 10  
print(f"Fibonacci sequence of {count} numbers:", fibonacci(count))

#5 : Calculate binary representation of given number
z = int(input("Please enter an integer: "))
binary = [ ] 
while z > 0:
  remainder = z % 2 
  binary.append(remainder) 
  z //= 2 
print("The wrong order", binary)
binary = binary[::-1] 
print("The binary representation of", z ,"is", binary)


```
